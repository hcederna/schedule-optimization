## Introduction

We selected our two best optimized volleyball conference schedules based on a composite scoring metric––__[schedule score](./Review.ipynb#score_metrics)__––that quantifies the following cost and team experience goals:

- __Total Cost__

> __Minimize total cost__ to free additional capital for academic-related expenditures per the NESCAC's emphasis on academics over athletics. __[[1]](http://www.nescac.com/about/mission_statement "NESCAC Mission Statement")__

- __Team Home Games__

> __Ensure each team has at least one home game__ and __evenly distributes additional team home games__ to equitably allocate the home-court advantage between teams.

- __Team Travel Duration__

> __Minimize and evenly distribute team travel duration__ to minimize and equitably allocate the travel-related fatigue and lost study hours between teams.

- __Team Overnights \*__

> __Minimize and evenly distribute team overnights__ to minimize and equitably allocate nights spent in hotels causing interruptions to player routine between teams.

__\*__ _A team overnight (hotel stay) occurs when a team plays a match on Friday and Saturday without returning home in between or when a team plays a match farther than a specified distance from home regardless of the day._

## Single Schedule Comparison
 
When compared to the previous NESCAC volleyball conference schedules, our `s0_1_3` optimized schedule saves money without sacrificing team experience, and our `s5_2_3` optimized schedule saves money with a slight sacrifice to the even distribution of overnights between teams.

Let's take a closer look.

![](results/single/figures/total_cost_($)_comparison.png)

Our optimized schedules __reduce the total cost by up to 27\%__ (\~20,400 USD) compared to the most expensive 2016 season, by at least 13\% (\~8,500 USD) compared to the least expensive 2010 season, and by 21\% (\~15,000 USD) on average compared to the previous 10 seasons. __[[2]](./results/single/figures/total_cost_($)_change.png)__ __[[3]](./results/single/figures/total_cost_percent_change.png)__ 

![](results/single/figures/team_home_game_comparison.png)

Both optimized schedules __ensure each team has at least 1 home game__ with 1 and 2 minimum team home games for the `s0_1_3` and `s5_2_3` schedules, respectively.

Both optimized schedules also __distribute team home games at least as evenly as the previous NESCAC schedules__ based on the __[team home game fairness index](./Review.ipynb#fairness_index)__ where a value of 100 indicates that every team has the same number of home games––we have 100% fairness. As team home game disparity increases, the team home game fairness index decreases towards the minimum possible value of 0. 

![](results/single/figures/team_duration_in_hours_comparison.png)

Our optimized schedules __reduce the average number of hours spent traveling per team by up to 22%__ compared to the most travel-heavy 2016 season and by at least 1% compared to the least travel-heavy 2011 season. __[[4]](./results/single/figures/team_duration_in_hours_average_percent_change.png)__
 
Both optimized schedules also __distribute team travel duration with a level of fairness in line with the previous NESCAC schedules__ (as quantified by the team travel duration fairness index).

![](results/single/figures/team_overnight_comparison.png)

Our optimized schedules __reduce the average number of overnights per team by up to 31%__ compared to the most overnight-heavy 2016 and 2017 seasons and by at least 22% compared to the least overnight-heavy 2009 and 2010 seasons. __[[5]](./results/single/figures/team_overnight_average_percent_change.png)__
 
The `s0_1_3` optimized schedule also __distributes team overnights with a level of fairness in line with the previous NESCAC schedules__ (as quantified by the team overnight fairness index). The `s5_2_3` optimized schedule, however, __distributes overnights between teams less fairly than the previous NESCAC schedules__. __[[6]](./results/single/figures/team_overnight_swarmplot.png)__

## Conclusion

The `s0_1_3` optimized volleyball conference schedule reduces total cost by 21% on average compared to the previous 10 NESCAC volleyball conference seasons without sacrificing team experience. A win all around.

The `s5_2_3` optimized volleyball conference schedule increases the minimum number of team home games from 1 in the `s0_1_3` optimized schedule to 2. It reduces total cost by 21% on average compared to the previous 10 NESCAC volleyball conference seasons while improving or at least maintaining (when compared with previous NESCAC seasons) the team home game experience, team travel experience, and number of team overnights. The increase in minimum team home games in the `s5_2_3` schedule, however, comes with the cost of decreasing the fairness of the team overnight distribution––a reasonable tradeoff if we prioritize the minimum number of team home games over the even distribution of team overnights.

### _A Final Note on the `s0_1_3` Schedule_

Assuming each team has 14 players and 2 coaching staff, the decrease in total travel time in the `s0_1_3` schedule translates to __saving 728 study hours and 104 coach hours per season__ compared to the most travel-heavy 2016 season. __[[7]](./results/single/figures/total_travel_duration_in_hours_comparison.png)__ __[[8]](./results/single/figures/total_travel_duration_in_hours_change.png)__

The decrease in total number of overnights in the `s0_1_3` schedule translates to __saving 128 nights spent away from home in a hotel per season__ compared to the most overnight-heavy 2016 and 2017 seasons. __[[9]](./results/single/figures/total_overnights_comparison.png)__ __[[10]](./results/single/figures/total_overnights_change.png)__
