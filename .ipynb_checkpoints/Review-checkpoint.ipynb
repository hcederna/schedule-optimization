
### Volleyball Conference Schedule

The NESCAC volleyball conference schedule occurs every fall season with the following requirements:

- __Eleven teams compete over five weeks__
- __Each team plays every other team one-time__
- __Each team plays no more than three games per week__
- __Each game occurs at home, away, or a neutral venue (neither home nor away)__

Beyond these mandates, there are several additional factors to consider. Ideally, a NESCAC volleyball conference schedule would:

- __Total Cost__

> __Minimize total cost__ to free additional capital for academic-related expenditures per the NESCAC's emphasis on academics over athletics.

- __Team Home Games__

> __Ensure each team has at least one home game__ and would __evenly distribute additional team home games__ to equitably allocate the home-court advantage between teams.

- __Team Travel Duration__

> __Minimize and evenly distribute team travel duration__ to minimize and equitably allocate the travel-related fatigue and lost study hours between teams.

- __Team Overnights \*__

> __Minimize and evenly distribute team overnights__ to minimize and equitably allocate the interruptions to player routine between teams.

__\*__ A team overnight (hotel stay) occurs when a team plays a match on Friday and Saturday without returning home in between or when a team plays a match farther than a specified distance from home regardless of the day.

### Scrape Schedules

We ultimately want to analyze the NESCAC volleyball conference schedules from the 2009 season (my first year playing at Bowdoin) through the most recent 2019 season, but first, we need to access these schedules.

We utilize the `BeautifulSoup` library to scrape each NESCAC volleyball schedule from the NESCAC website into a `pandas` DataFrame. We then for each year save a cleaned conference schedule DataFrame and a team conference schedule DataFrame to a comma-separated value (CSV) file.

For a more in-depth explanation of this process, check out my blog post, __[Scrape a Website Table into a Pandas DataFrame](../blog/scrape/Scrape.ipynb)__.

Let's get scraping.

```python
from src.configuration import *
from src.scrape_schedule import *
from src.helpers import create_team_schedule_df


# List of schedule years to scrape
years = list(range(2009, 2020))

# List of saved NESCAC website HTML files
html_filenames = ["hard_data/" + str(year) + "_nescac_schedule.html" for year in years]

# Iterate over each schedule year to scrape
for i in range(len(years)):

    year = years[i]
    
    print("Scraping {} NESCAC conference schedule from NESCAC website ...".format(year))

    # Scrape data from saved NESCAC website HTML file(s)
    html_file = html_filenames[i]
    schedule_df = scrape_clean_create_nescac_schedule_df(
        year=year,
        html_file=html_file, 
        venues_mapping_file=VENUES_MAPPING_FILENAME)
    
    print("Saving conference schedule to CSV ...")
    schedule_filename = "%s%s_nescac_schedule.csv" %(NESCAC_SCHEDULE_PATH, year)
    schedule_df.to_csv(schedule_filename)
    print("... Saved as \"{}\"".format(schedule_filename))

    print("Preview:")
    display(schedule_df.head())
    print()
    
    print("Creating {} NESCAC team conference schedule ...".format(year))
    team_schedule_df = create_team_schedule_df(schedule_df)
    
    print("Saving team conference schedule to CSV ...")
    team_schedule_filename = "%s%s_nescac_team_schedule.csv" %(NESCAC_SCHEDULE_PATH, year)
    team_schedule_df.to_csv(team_schedule_filename)
    print("... Saved as \"{}\"".format(team_schedule_filename))

    print("Preview:")
    display(team_schedule_df.head())
    print()

print("Process complete.\n")
```

#### Jain's fairness index

One metric to quantify fairness is Jain's fairness index __[[2]](https://www.cse.wustl.edu/~jain/papers/ftp/fairness.pdf)__ calculated as:

$\begin{align*} \qquad f(x_1, x_2, \cdots, x_n) = 100 * \frac{(\sum_{i=1}^{n} x_i)^2}{n * \sum_{i=1}^{n} (x_i^2)}\end{align*}$

where $n$ in this scenario is the number of teams and $x_i$ is team $i$'s number of home games.

We use Jain's fairness index because it is:
- independent of physical units
- independent of scale
- within the range [0, 100]
- continuous such that a small change in $x_i$ leads to a small change in $f(x)$

A fairness index value of 100 indicates all $x_i$'s are equal (every team has the same number of home games)we have 100% fairness. As disparity increases, the fairness index decreases towards the minimum possible value of 0.

##### Example fairness index calculation

Let's take a closer look at the formula for Jain's fairness index and calculate the fairness index for team home games in the 2010 season.

We can rewrite the formula for Jain's fairness index __[[3]](../blog/derivation/Derivation.ipynb)__ as:

$\begin{align*} \qquad f(x_1, x_2, \cdots, x_n) = 100 * \frac{1}{n} * \sum_{i=1}^{n} (\frac{x_i}{x_f}) \end{align*}$

with $x_f$ representing the fairness allocation mark:

$\begin{align*} \qquad x_f = \frac{\sum_{i=1}^{n} (x_i^2)}{\sum_{i=1}^{n} x_i} \end{align*}$

In this framework, each team's home game allocation is only $x_i$/$x_f$ fair. An $x_i$ > $x_f$ indicates a favored team home game distribution while an $x_i$ < $x_f$ indicates a disadvantaged team home game distribution. The overall fairness index is then just the average fairness for all teams.

In the 2010 schedule, there was 1 team with 1 home game, 3 teams with 2 home games, and 7 teams with 3 home games. Let's calculate the fairness allocation mark for this team home game distribution:

$\begin{align*} \qquad x_f = \frac{(1 \cdot 1^2) + (3 \cdot 2^2) + (7 \cdot 3^2)}{(1 \cdot 1) + (3 \cdot 2) + (7 \cdot 3)} = \frac{(1 \cdot 1) + (3 \cdot 4) + (7 \cdot 9)}{(1 \cdot 1) + (3 \cdot 2) + (7 \cdot 3)} = 2.714 \end{align*}$


We can then use the $x_i$/$x_f$ formula to determine:

- Having 1 home game is `1 / 2.714` = 36.84% fair
- Having 2 home games is `2 / 2.714` = 73.69% fair
- Having 3 home games is `3 / 2.714` = 110.54% fair
    
The overall fairness index for the 2010 team home game distribution is therefore:

$\begin{align*} \qquad f(x) = 100 \cdot \frac{1}{11} \cdot [(1 \cdot 0.3684) + (3 \cdot 0.7369) + (7 \cdot 1.1054)] = 93.79 \end{align*}$

For comparability, we create scores for each of the schedule metrics that we deem essential. We calculate a team home game score (quantifying the goal to distribute team home games evenly) as the inverse team home game fairness index rescaled to a [0, 100] range. 

We use the inverse fairness index because we make all scores inverse scores where small values represent the best results and high values represent the worst results. A score value of 0 is the best. A score value of 100 is the worst. Accordingly, to have a score that quantifies maximizing the team home game fairness index, we need to minimize the inverse team home game fairness index. 

We rescale each data point used in a score metric to a [0, 100] range to ensure all score components are weighted evenly in a composite score. This ultimately prevents a schedule's total cost from drowning out other factors in the composite schedule score. 

#### A quick note on rescaling

Each data point used in a score metric is rescaled to a range of [0, 100] using the formula:

$ \quad X' = 100*\frac{X \:-\: X_{min}}{X_{max} \:-\:X_{min}} $

Because of this rescaling, the score values are relative to the other schedules being analyzed. The `x_max` and `x_min` for the previous NESCAC schedules will be different than the `x_max` and `x_min` for the optimized schedules which will be different from the `x_max` and `x_min` for the previous NESCAC schedules and optimized schedules combined. The same schedule in each of these groups might have different score values. Accordingly, before we compare scores, we need to ensure the score calculations include all of the schedules we wish to compare.

### Score Metrics <a id='score_metrics'></a>


Now that we reviewed how the previous NESCAC volleyball conference schedules performed with respect to:
- Total cost
- Team home games
- Team travel duration
- Team overnights

Let's create a handful of composite score metrics to represent each schedule more holistically. 

We create a team experience score as the average of the team score components:

$
\quad Team\:Experience\:Score = \\
\qquad \frac{1}{5} * Inverse\:Team\:Home\:Game\:Fairness\:Index\:Rescaled + \\
\qquad \frac{1}{5} * Inverse\:Team\:Duration\:in\:Hours\:Fairness\:Index\:Rescaled + \\
\qquad \frac{1}{5} * Team\:Duration\:in\:Hours\:Average\:Rescaled + \\
\qquad \frac{1}{5} * Inverse\:Team\:Overnight\:Fairness\:Index\:Rescaled + \\
\qquad \frac{1}{5} * Team\:Overnight\:Average\:Rescaled
$

We create a schedule score that weighs the total cost and team score components based on the schedule score total cost weight (`TCW`) specified in the analyze schedule section of __[configuration.py](./src/configuration.py)__:

$
\quad Schedule\:Score = \\
\qquad TCW * Total\:Cost\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Inverse\:Team\:Home\:Game\:Fairness\:Index\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Inverse\:Team\:Duration\:in\:Hours\:Fairness\:Index\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Team\:Duration\:in\:Hours\:Average\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Inverse\:Team\:Overnight\:Fairness\:Index\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Team\:Overnight\:Average\:Rescaled
$

Remember that all scores are inverse scores where small values represent the best results, and high values represent the worst results. A score value of 0 is the best. A score value of 100 is the worst.
