```python
# Closer look at team overnight dict values for optimized
# schedules in optimized quad + worst performing in nescac quad
# based on overnight fairness
columns = [
    'Abbreviated Schedule ID',
    'Schedule ID 0',
    'Schedule ID 1',
    'Schedule ID 2',
    'Schedule ID 3',
]

# load necessary info
optimized_schedule_analysis_df = pd.read_pickle(OPTIMIZED_SCHEDULE_ANALYSIS_FILENAME)
nescac_schedule_analysis_df = pd.read_pickle(NESCAC_SCHEDULE_ANALYSIS_FILENAME)
quad_schedule_comparison_df = pd.read_pickle(QUAD_SCHEDULE_COMPARISON_FILENAME)

# extract relevant optimized quads + nescac quad to compare to
_ = quad_schedule_comparison_df.iloc[selected_idxs][columns].append(quad_schedule_comparison_df[quad_schedule_comparison_df['Abbreviated Schedule ID'] == '2009_10_11_12'][columns])
# display(_)

# schedule ID columns
sched_id_cols = [col for col in _.columns if "Abbreviated" not in col]


# create data dict to hold scheds info + team overnight values
data = {
    'Abbreviated Schedule ID': [],
    'Schedule # in Quad': [],
    'Schedule ID': [],
    'Team Overnight Dict': [],

}
# iterate over optimized quads
for idx, row in _.iterrows():
    
    abb_sched_id = row['Abbreviated Schedule ID']
    
    for sched_id_col in sched_id_cols:
                
        data['Schedule ID'].append(row[sched_id_col])
        data['Schedule # in Quad'].append(sched_id_col[-1])
        if "nescac" in row[sched_id_col]:
            team_overnight_dict = nescac_schedule_analysis_df[nescac_schedule_analysis_df['Schedule ID'] == row[sched_id_col]]['Team Overnight Dict'].item()
            data['Abbreviated Schedule ID'].append(row[sched_id_col][2:8])
        else:
            team_overnight_dict = optimized_schedule_analysis_df[optimized_schedule_analysis_df['Schedule ID'] == row[sched_id_col]]['Team Overnight Dict'].item()
            data['Abbreviated Schedule ID'].append(abb_sched_id[:3] + "_sch_" + sched_id_col[-1])

            
        data['Team Overnight Dict'].append(team_overnight_dict)

team_overnight_df = pd.DataFrame(data)
display(team_overnight_df)
```

### Team Overnight + Duration --> Percent Change Columns

```python
from src.configuration import WEEKEND_OVERNIGHT_THRESHOLD, DISTANCE_OVERNIGHT_THRESHOLD

print(f"WEEKEND_OVERNIGHT_THRESHOLD = {WEEKEND_OVERNIGHT_THRESHOLD}")
print("""
    \tMaximum number of miles to drive one-way without staying for an
    \tovernight between Friday and Saturday matches (recommended = 53
    \t==> max one-way distance with duration < 1 hour)
    """)
      
print(f"DISTANCE_OVERNIGHT_THRESHOLD = {DISTANCE_OVERNIGHT_THRESHOLD}")
print("""
    \tMaximum number of miles to drive one-way without staying for an
    \tovernight (recommended = 153 ==> max one-way distance with duration
    \t< 3 hours)
    """)


for filename in filenames[:2]:
    
    print("########################################")
    print(f"# {filename} #")
    print("########################################")

    schedule_df = pd.read_csv(f"results/quad/optimized_schedules/{filename}", index_col=0)

    schedule_df['Home Game'] = schedule_df['Team 1'] == schedule_df['Location']
    schedule_df['Team 1 Name'] = schedule_df['Team 1']
    schedule_df['Location Name'] = schedule_df['Location']
    schedule_df['Team 1'] = schedule_df['Team 1'].map(name_to_id_mapping)
    schedule_df['Location'] = schedule_df['Location'].map(name_to_id_mapping)

    schedule_df['Team 1 Travel Distance (mi)'] = None
    schedule_df['Team 1 Travel Distance (hhmm)'] = None
    for idx, row in schedule_df.iterrows():
        schedule_df['Team 1 Travel Distance (mi)'][idx] = venue_one_way_distance_matrix.iloc[row['Team 1'], row['Location']]
        schedule_df['Team 1 Travel Distance (hhmm)'][idx] = venue_one_way_duration_matrix_hhmm.iloc[row['Team 1'], row['Location']]


    schedule_df['Distance Overnight'] = False
    schedule_df['Weekend Overnight'] = False
    for team_week, team_week_df in schedule_df.groupby(['Team 1', 'Week']):

        team_week_df = team_week_df.reset_index(drop=False)
        curr_team = team_week[0]

        # Iterate over team 1 matches for the week
        for idx, row in team_week_df.iterrows():
            current_match = team_week_df.loc[idx]
            schedule_df_idx = current_match['index']

            # Check for a distance overnight
            if is_distance_overnight(current_match, venue_one_way_distance_matrix, DISTANCE_OVERNIGHT_THRESHOLD):
                schedule_df['Distance Overnight'][schedule_df_idx] = True

            # Check for a weekend overnight
            if idx > 0:
                previous_match = team_week_df.loc[idx - 1]

                if is_weekend_overnight(previous_match, current_match, venue_one_way_distance_matrix, WEEKEND_OVERNIGHT_THRESHOLD):

                    # Count weekend overnight
                    schedule_df['Weekend Overnight'][schedule_df_idx] = True

                    # Remove distance overnight because have weekend overnight ==> already staying overnight
                    schedule_df['Distance Overnight'][schedule_df_idx] = False

                    break

    # Display results                
    columns = [
        'Week',
        'Day (Str)',
        'Time Slot',
        'Team 1 Name',
        'Team 2',
        'Location Name',
        'Home Game',
        'Team 1 Travel Distance (mi)',
        'Team 1 Travel Distance (hhmm)',
        'Distance Overnight',
        'Weekend Overnight'
    ]

    # for team_1_group_name, team_1_group in _.groupby('Team 1'):

    #         display(team_1_group[columns])
    #         print(single_schedule_comparison_df['Team Overnight Dict'][0][team_1_group_name])

    optimized_schedule_analysis_df = pd.read_pickle(OPTIMIZED_SCHEDULE_ANALYSIS_FILENAME)

    for team_1, team_1_df in schedule_df.groupby('Team 1'):

        print(f"TEAM {team_1}")
        _ = optimized_schedule_analysis_df[optimized_schedule_analysis_df['Schedule ID'] == filename[:-4]]
        print(f"   overnight = {_['Team Overnight Dict'].item()[team_1]}")
        print(f"   distance = {_['Team Distance Dict'].item()[team_1]}")

        display(team_1_df)

    #     for week, week_df in team_1_df.groupby('Week'):

    #         display(week_df[columns])
```
