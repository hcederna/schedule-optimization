# A Closer Look at the Optimized Single Schedules

## Optimized Schedule `s0_1_3`

```python
import pandas as pd

filename = '15_2335_optimized_team_schedule_0.csv'
```

### Full Schedule

```python
# toggle/untoggle display lines to review full schedule or full schedule grouped by week
print(f"\n{filename.replace('team_', '')}\n")

schedule_df = pd.read_csv(f"results/single/optimized_schedules/{filename.replace('team_', '')}", index_col=0)

# display(schedule_df)

# review schedule grouped by week
for week, week_df in schedule_df.groupby('Week'):
    display(week_df)
```

### Team Schedule

```python
# toggle/untoggle display lines to review team schedules grouped by team or team and week    
print(f"\n{filename}\n")

schedule_df = pd.read_csv(f"results/single/optimized_schedules/{filename}", index_col=0)

for team_1, team_1_df in schedule_df.groupby('Team 1'):
    print(team_1)
    # review team schedule grouped by team        
    display(team_1_df)

#     # review team schedule grouped by team and week
#     for week, week_df in team_1_df.groupby('Week'):
#         display(week_df)
```
