# Compare Single Optimized & Previous NESCAC Volleyball Conference Schedules

This notebook picks up where the __[Optimize the NESCAC Volleyball Conference Schedule](./Optimize.ipynb)__ notebook left off. It is worthwhile to at least skim the previous notebook before proceeding.


## Introduction

We selected our optimized NESCAC volleyball conference schedules based on the following goals:

- __Minimize total cost__
- __Ensure each team has at least one home game__
- __Evenly distribute additional team home games__ : maximize team home game fairness index
- __Minimize team travel duration__ : minimize team travel duration average
- __Evenly distribute team travel duration__ : maximize team duration in hours fairness index
- __Minimize team overnights__ : minimize team overnight average
- __Evenly distribute team overnights__ : maximize team overnight fairness index

We quantified these goals in a schedule score that weighs the total cost and team score components based on the schedule score total cost weight (`TCW`) specified in the analyze schedule section of __[configuration.py](./src/configuration.py)__:

- __Schedule Score =__

$
\qquad TCW * Total\:Cost\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Inverse\:Team\:Home\:Game\:Fairness\:Index\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Inverse\:Team\:Duration\:in\:Hours\:Fairness\:Index\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Team\:Duration\:in\:Hours\:Average\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Inverse\:Team\:Overnight\:Fairness\:Index\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Team\:Overnight\:Average\:Rescaled
$

Remember that all scores are inverse scores where small values represent the best results, and high values represent the worst results. A score value of 0 is the best. A score value of 100 is the worst.

#### A quick note on rescaling

Each data point used in the schedule score metric is rescaled to a range of [0, 100] using the formula:

$ \quad X' = 100*\frac{X - X_{min}}{X_{max} - X_{min}} $

Because of this rescaling, the score values are relative to the other schedules being analyzed. The `x_max` and `x_min` for the previous NESCAC schedules are different than the `x_max` and `x_min` for the optimized schedules, which are different from the `x_max` and `x_min` for the previous NESCAC schedules and optimized schedules combined. The same schedule in each of these groups might have different score values. Accordingly, before we compare scores, we need to ensure the score calculations include all of the schedules we wish to compare.

```python
name_to_id_mapping = create_venue_id_mapping_dicts_from_file(VENUES_MAPPING_FILENAME)[0]
name_to_id_mapping
```

```python
venue_one_way_distance_matrix = pd.read_csv(DISTANCE_MATRIX_FILENAME, index_col=0)
venue_one_way_distance_matrix
```

```python
venue_one_way_duration_matrix_hhmm = pd.read_csv(DURATION_MATRIX_HHMM_FILENAME, index_col=0)
venue_one_way_duration_matrix_hhmm
```

```python
schedule_index = 1

_ = pd.read_csv(f"results/single/optimized_schedules/{single_schedule_comparison_df['Schedule ID'][schedule_index]}.csv", index_col=0)

_['Home Game'] = _['Team 1'] == _['Location']
_['Team 1 Name'] = _['Team 1']
_['Location Name'] = _['Location']
_['Team 1'] = _['Team 1'].map(name_to_id_mapping)
_['Location'] = _['Location'].map(name_to_id_mapping)
_['Team 1 Travel Distance (mi)'] = None
_['Team 1 Travel Distance (hhmm)'] = None

for idx, row in _.iterrows():
    
    _['Team 1 Travel Distance (mi)'][idx] = venue_one_way_distance_matrix.iloc[row['Team 1'], row['Location']]
    _['Team 1 Travel Distance (hhmm)'][idx] = venue_one_way_duration_matrix_hhmm.iloc[row['Team 1'], row['Location']]


_['Distance Overnight'] = False
_['Weekend Overnight'] = False

# Maximum number of miles to drive one-way without staying for an
# overnight between Friday and Saturday matches
# Recommended = 53
# ==> max one-way distance with duration < 1 hour
weekend_overnight_threshold = 53

# Maximum number of miles to drive one-way without staying for an
# overnight
# Recommended = 153
# ==> max one-way distance with duration < 3 hours
distance_overnight_threshold = 153

# # Temporarily add a week column to the NESCAC team schedule DataFrame
# if "Week" not in _.columns:

#     _['Date'] = pd.to_datetime(_['Date'])
#     _['Week'] = _['Date'].dt.strftime('%U')

# Group team schedule DataFrame by team 1 and week
for team_week, team_week_df in _.groupby(['Team 1', 'Week']):

    team_week_df = team_week_df.reset_index(drop=False)
    curr_team = team_week[0]

    # Iterate over team 1 matches for the week
    for idx, row in team_week_df.iterrows():

        current_match = team_week_df.loc[idx]
        _idx = current_match['index']

        # Check for a distance overnight
        if is_distance_overnight(current_match, venue_one_way_distance_matrix, distance_overnight_threshold):

            _['Distance Overnight'][_idx] = True

#             distance_overnight = True

        # Check for a weekend overnight
        if idx > 0:

            previous_match = team_week_df.loc[idx - 1]

            if is_weekend_overnight(previous_match, current_match, venue_one_way_distance_matrix, weekend_overnight_threshold):

                # Count weekend overnight
                _['Weekend Overnight'][_idx] = True

#                 team_overnight_dict[curr_team] += 1

                # Remove distance overnight because have weekend overnight ==> already staying overnight
                _['Distance Overnight'][_idx] = False

#                 distance_overnight = False

                break
```
