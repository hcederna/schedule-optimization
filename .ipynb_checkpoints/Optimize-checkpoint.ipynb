# Optimize the NESCAC Volleyball Conference Schedule

This notebook picks up where the __[Review Previous NESCAC Volleyball Conference Schedules](./Review.ipynb)__ notebook left off. It is worthwhile to at least skim the previous notebook before proceeding.

## Introduction

### Problem formulation

Let's review the NESCAC volleyball conference schedule requirements:

- __Eleven teams compete over five weeks__
- __Each team plays every other team one-time__
- __Each team plays no more than three games per week__
- __Each game occurs at home, away, or a neutral venue (neither home nor away)__

For the sake of this optimization, let's add the additional requirements:

- __Each location hosts no more than one match per time slot__ 
- __Matches are evenly distributed throughout the season__

Beyond these mandates, there are several additional factors to consider.

- __Minimize total cost__
- __Ensure each team has at least one home game__
- __Evenly distribute additional team home games__ : maximize team home game fairness index
- __Minimize team travel duration__ : minimize team travel duration average
- __Evenly distribute team travel duration__ : maximize team duration in hours fairness index
- __Minimize team overnights__ : minimize team overnight average
- __Evenly distribute team overnights__ : maximize team overnight fairness index

These additional factors are combined in a schedule score that weighs the total cost and team score components based on the schedule score total cost weight (`TCW`) specified in the analyze schedule section of __[configuration.py](./src/configuration.py)__:

- __Schedule score__

$
\qquad = TCW * Total\:Cost\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Inverse\:Team\:Home\:Game\:Fairness\:Index\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Inverse\:Team\:Duration\:in\:Hours\:Fairness\:Index\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Team\:Duration\:in\:Hours\:Average\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Inverse\:Team\:Overnight\:Fairness\:Index\:Rescaled + \\
\qquad (1 - TCW) * \frac{1}{5} * Team\:Overnight\:Average\:Rescaled
$

Remember that all scores are inverse scores where small values represent the best results, and high values represent the worst results. A score value of 0 is the best. A score value of 100 is the worst.

#### A quick note on rescaling

Each data point used in the schedule score metric is rescaled to a range of [0, 100] using the formula:

$ \quad X' = 100*\frac{X - X_{min}}{X_{max} - X_{min}} $

Because of this rescaling, the score values are relative to the other schedules being analyzed. The `x_max` and `x_min` for the previous NESCAC schedules are different than the `x_max` and `x_min` for the optimized schedules, which are different from the `x_max` and `x_min` for the previous NESCAC schedules and optimized schedules combined. The same schedule in each of these groups might have different score values. Accordingly, before we compare scores, we need to ensure the score calculations include all of the schedules we wish to compare.

### Literature review

After reviewing the literature on sports scheduling problems, we can frame optimizing the NESCAC volleyball conference schedule as a variation/customization of the traveling tournament problem originally outlined by Easton et al. (2001) __[[1]](https://mat.tepper.cmu.edu/trick/ttp.pdf)__. In this framework, the most straightforward optimization would use constraint-based integer programming. We would create a decision variable: 

$ \quad
X_{ijkl} =
\begin{cases}
    1 & \text{if team } i \text{ plays team } j \text{ at location } l \text{ in time slot } k \\
    0 & \text{otherwise}\\
\end{cases}
$

And apply constraints to ensure we satisfy the schedule requirements. <a id='optimization_strategy'></a>We would then solve the problem with an objective function that minimizes the total distance traveled. __\*__ 

Sounds great, but we quickly run into a problem. The optimization takes an infeasible amount of time to solve. For context, Easton et al. (2003)  __[[2]](http://mat.tepper.cmu.edu/TOURN/ttp_ipcp.pdf)__ proposes a solution to the 8 team traveling tournament problem (previously left unsolved in Easton et al. (2001) despite days of computation time) using a combination of integer programming and constraint programming. The optimization took 20 processors approximately 4 days to output a solution. Now, technology has advanced since 2003, but our NESCAC volleyball conference scheduling problem is more complex and involves 11 teams rather than 8. We need to consider an alternative, more practical approach. 

__\*__ _We discuss __[why we want to minimize the total distance traveled](Review.ipynb#optimization_strategy)__ in the previous notebook._

## Proposed Solution

Using a novel combination of clustering and integer programming, we can optimize the NESCAC volleyball conference schedule in a reasonable amount of time. Here are the steps:

 1\. __Cluster the NESCAC schools__ 

> We reduce the solution space by grouping the NESCAC schools into clusters. This allows us to utilize integer programming with a reduced number of NESCAC groups rather than all 11 of the NESCAC teams. We use the k-means clustering algorithm to create multiple potential clusterings and compile the results in a clusterings information DataFrame.


2\. __Optimize a conference schedule__ <a id='optimization_strategy'></a>

> Using a clustering from the previous step, we apply constraint-based integer programming to meet schedule requirements and minimize the total distance traveled at the group-level. We then utilize enumeration to transform the group-level schedule to a team-level schedule while ensuring we evenly distribute home games.

3\. __Survey clusterings and constraint parameters__

> We create several clusterings in step 1, and we use a tuple of values in step 2 to ensure home games are evenly distributed. In this step, we create multiple NESCAC volleyball conference schedules for each combination of clustering and tuple of values. We compile the average results for each combination in a parameter survey DataFrame.

4\. __Create multiple NESCAC volleyball conference schedule using the best parameter settings__

> We use the best of the acceptable parameter combinations from step 3 to create multiple optimized NESCAC volleyball conference schedules. We analyze each of the optimized schedules and compile the schedule and analysis information in an optimized schedule analysis DataFrame. 

5\. __Select and save the best optimized NESCAC volleyball conference schedules__

> We select and save the best single optimized schedules based on the defined scoring metric––schedule score. We also select and save the best groups of two and four schedules, accounting for the average schedule score and match location coverage and changes.

Let's take a closer look.

### Cluster the NESCAC schools

The k-means algorithm classifies points into a specified number `k` of clusters using the following steps:

1. Randomly initialize `k` cluster centroids
2. Calculate the distance from each point to each centroid using a specified distance metric
3. Assign each point to the closest centroid
4. Update the centroid location as the average of the points assigned to the centroid's cluster
5. Repeat steps 2 through 4 until cluster assignments and centroid updates no longer change

We identify the location of each NESCAC school using latitude and longitude values––geographic coordinates. Accordingly, we need to pay special attention to our distance metric in k-means.

The Python libraries `SciPy` and `SciKit-Learn` implement the k-means algorithm using Euclidean distance––the straight-line distance between two points. Euclidean distance accurately represents the distance in the Cartesian coordinate system. In the geographic coordinate system, however, the earth is round. A straight-line path between two geographic coordinates is not an accurate representation of the distance (especially over greater distances). Accordingly, we will:

- Convert the NESCAC school geographic coordinates to Cartesian coordinates and use a Euclidean distance metric
- Use the NESCAC school geographic coordinates and a Haversine distance metric 
> Haversine calculates the distance between latitude and longitude locations on the surface of the assumed spherical earth. The earth is actually ellipsoidal, and the spherical assumption can lead to an error of up to 0.5% __[[3]](https://geopy.readthedocs.io/en/stable/)__.
- Use the NESCAC school geographic coordinates and a Vincenty distance metric
> Vincenty calculates the distance between latitude and longitude locations on the surface of the earth using the most accurate ellipsoidal model of the earth––WGS-84 __[[4]](https://geopy.readthedocs.io/en/stable/)__.
    
To utilize the Haversine and Vincenty distance metrics, we implement our own version of the k-means algorithm. We then run k-means using a combination of implementations (`scratch`, `scipy`, `sklearn`), coordinate types (`cartesian`, `geographic`), and distance metrics (`euclidean`, `haversine`, `vincenty`).

Our primary goal is to create:
- many possible NESCAC clusterings
- where each clustering groups the NESCAC schools into five clusters (one cluster for each week in the season)
- with each cluster composed of three or fewer schools (no more than the maximum of three games per week)

## Step 2 : Optimization

### Optimize a conference schedule

We optimize the NESCAC volleyball conference schedule with the following steps:

1. Using constraint-based integer programming, create a group-level conference schedule that meets the schedule requirements and minimizes total distance traveled. Outputs a group conference schedule as a list of `(i, j, k)` matches that indicate `group i plays at home against group j during week k`.

2. Using enumeration, assign match locations to evenly distribute the number of home games between teams in a group. Outputs a team location group schedule as a list of `(i, j, k, l)` matches that indicate `group i plays group j at team (a member of group i) location l during week k`.

3. Using enumeration, assign team from group i versus team from group j matchups to ensure each team in group i plays each team in group j and randomly assign time slots during the week for each matchup to occur. Outputs an optimized schedule as a list of `(i, j, k, l, m)` matches that indicate `team i plays team j at location l during week k in time slot m`.

Let's go through this step by step.

#### 1. Create a group conference schedule

Given the following:

$\quad n = \text{the number of groups}$

$\quad d(j,i) = \text{the one-way driving distance between the away group's centroid j and the home group's centroid i}$

$\quad w = \text{the number of weeks in the season}$

##### Decision Variables

The decision variables represent games in the schedule where:

$\quad 
X(i, j, k) = 
\begin{cases}
    1  & \text{if group } i \text{ plays at home against group } j \text{ in week } k {,} \\
    0  & \text{otherwise}\\
\end{cases}
$

Example:

$\quad X(1, 2, 0) = 0 $

$\qquad \quad $ Group 1 does not play at home against group 2 in week 0.

$\quad X(1, 2, 3) = 1 $

$\qquad \quad $ Group 1 plays at home against group 2 in week 3.

##### Objective Function

The objective function minimizes the total distance traveled as follows:

$
\begin{align*}
\quad min \sum_{i=0}^{n-1} \sum_{j=0}^{n-1} \sum_{k=0}^{w-1} X(i, j, k) \cdot 2d(j, i)
\end{align*}
$

Example:

$\quad ( \cdots + X(1, 2, 0) \cdot 2d(2, 1) + \cdots) \quad$ where $\quad X(1, 2, 3) = 1$

$\qquad \quad $ Add to the objective function the roundtrip distance for group 2 to drive to group 1.

##### Constraints

To ensure the group schedule meets schedule requirements, we enforce the following constraints:

- __Each team plays every other team one-time__

>In order to meet the schedule requirement that each team plays every other team one time, we need to ensure that each group plays every other group (including itself) one time. We use the following constraint:

$\begin{align*} \qquad \sum_{k=0}^{w-1} X(i, j, k) + X(j, i, k) = 1, \quad \end{align*}$ for each $\quad i, j = 0, \cdots, n-1 : i \leq j$

$\quad $ Example:

$\qquad X(0, 1, 0) + X(1, 0, 0) + X(0, 1, 1) + X(1, 0, 1) + \cdots $ for all $ k \cdots = 1 $

$\qquad \qquad $ Group 0 plays at home against group 1 or group 0 plays away against group 1 only one time in the schedule.

$\quad $ Note that we add the $i \leq j$ component to avoid duplicate constraints:

$\qquad $ sum$ (\:X(0, 1, k) + X(1, 0, k)\:) = 1$ is needed

$\qquad $ sum$(\:X(1, 0, k) + X(0, 1, k)\:) = 1$ is a duplicate constraint

- __Each team plays no more than three games per week__

> We know from the clustering step that each group has no more than three teams. Accordingly, to meet the schedule requirement that each team plays no more than three games per week, we need to ensure that each group plays no more than one group per week. We use the following constraint:

$\begin{align*} \qquad \sum_{j=0}^{n-1} X(i, j, k) + X(j, i, k) \leq 1, \quad \end{align*}$ for each $\quad i = 0, \cdots, n-1, \quad k = 0, \cdots, w-1$

$\quad $ Example:

$\qquad X(3, 0, 0) + X(0, 3, 0) + X(3, 1, 0) + X(1, 3, 0) + \cdots $ for all $ j \cdots \leq 1 $

$\qquad \qquad $ Group 3 plays at home against an opponent or group 3 plays away against an opponent no more than one time in week 0. 

$\qquad \qquad $ Note that we have the $\leq$ because group 3 may not play in week 0 at all.

- __Each location hosts no more than one match per time slot__ 

>In order to meet the additional schedule requirement that each location hosts no more than one match per time slot, we need to ensure that each location hosts no more than one group per week. We use the following constraint:

$\begin{align*} \qquad \sum_{j=0}^{n-1} X(i, j, k) \leq 1, \quad \end{align*}$ for each $\quad i = 0, \cdots, n-1, \quad k = 0, \cdots, w-1$

$\quad $ Example:

$\qquad X(2, 0, 0) + X(2, 1, 0) + X(2, 2, 0) + \cdots $ for all $ j\cdots \leq 1 $

$\qquad \qquad $ Group 2 plays at home against any opponent no more than one time in week 0.

$\qquad \qquad $ Said another way, a group match occurs at location 2 no more than one time in week 0.

- __Matches are evenly distributed throughout the season__

> To meet the additional requirement that matches are evenly distributed throughout the season, we need to ensure that matches are evenly distributed throughout the weeks. We first calculate the total number of group matches throughout the season divided by the number of weeks. We then round up this number to determine the maximum number of matches per week to ensure even distribution. We use the following constraint:


$\begin{align*} \qquad \sum_{i=0}^{n-1} \sum_{j=0}^{n-1} X(i, j, k) \leq Maximum\:Weekly\:Matches, \quad \end{align*}$ for each $\quad k = 0, \cdots, w-1$

$\quad $ Example:

$\qquad X(0, 0, 0) + X(0, 1, 0) + X(0, 2, 0) + \cdots $ for all $ i,j\cdots \leq Maximum\:Weekly\:Matches $

$\qquad \qquad $ No more than the maximum number of weekly matches occurs in week 0.

- __Home games are evenly distributed between teams__

> To select for the additional factor that home games are evenly distributed between teams, we need to ensure that home games are evenly distributed between groups. We determine the acceptable minimum and acceptable maximum number of home games that result in the best optimized schedules in a future parameter survey. We then use these values in the following constraints:

$\begin{align*} \qquad \sum_{j=0}^{n-1} \sum_{k=0}^{w-1} X(i, j, k) \geq Acceptable\:Minimum, \quad \end{align*}$ for each $\quad i = 0, \cdots, n-1$

$\begin{align*} \qquad \sum_{j=0}^{n-1} \sum_{k=0}^{w-1} X(i, j, k) \leq Acceptable\:Maximum, \quad \end{align*}$ for each $\quad i = 0, \cdots, n-1$

$\quad $ Example:

$\qquad Acceptable\:Minimum \leq X(0, 0, 0) + X(0, 1, 0) + X(0, 2, 0) + \cdots $ for all $ j,k\cdots \leq Acceptable\:Maximum $

$\qquad \qquad $ Group 0 plays at home against any opponent in any week between the acceptable minimum and acceptable maximum number of times.

##### Solution

We solve the constraint-based integer program to create a group schedule that minimizes total distance traveled subject to the constraints. The output is a group conference schedule as a list of `(i, j, k)` matches that indicate `group i plays at home against group j during week k`.

#### 2. Assign match locations

Our next step is to take the group schedule of `(i, j, k)` matches where `group i plays at home against group j during week k` and assign a team (a member of group i) location l for the group match. We want to assign locations such that the number of home games are evenly distributed between teams in a group.

For each group, we create a list of possible location combinations for group home matches. We then select the location combination that minimizes the range of team home games within a group.

Example:

$ \quad $ Group 1 Home Matches $ = [(1, 2, 3), (1, 0, 0), (1, 1, 1), (1, 3, 4), (1, 4, 2)]$

$ \quad $ Group 1 Teams $ = [7, 9]$

$ \quad $ Location Combinations $ = [(7, 9, 7, 9, 7), (9, 9, 9, 9, 9), \cdots ]$

$ \quad $ Winning Combination $ = (7, 9, 7, 9, 7) \quad $

$ \qquad $ with Home Game Range = 1 $ \quad $ (3 games @ location 7, 2 games @ location 1 ==> 3 - 2 = 1)

We output a team location group schedule as a list of `(i, j, k, l)` matches that indicate `group i plays group j at team (a member of group i) location l during week k`.

Example:

$ \quad (1, 2, 3, 7)$

$ \qquad \quad $ Group 1 plays group 2 at the venue of team 7 in week 3.

####  3. Assign team matches & time slots

We are now ready to run our final step––assigning team from group i versus team from group j matchups to ensure that each team in group i plays each team in group j. 

For each match in the team location group schedule, we create a list of matchups––team from group i versus team from group j. We then randomly shuffle the matchups and assign each matchup to a time slot.

Example:

$ \quad $ Team Location Group Schedule Match $ = (1, 2, 3, 7)$

$ \quad $ Group 1 Teams $ = [7, 9]$

$ \quad $ Group 2 Teams $ = [0, 8]$

$ \quad $ Matchups $ = [(7, 0), (7, 8), (9, 0), (9, 8)]$

$ \quad $ Shuffle Matchups $ = [(7, 8), (9, 8), (9, 0), (7, 0)]$

We output a schedule as a list of `(i, j, k, l, m)` matches that indicate `team i plays team j at location l during week k in time slot m`.

Example:

$ \quad (7, 8, 3, 7, 0) $

$ \qquad \quad $ Team 7 plays against team 8 at the venue of team 7 in week 3 and time slot 0.

#### Create optimized conference schedule

We put these three steps together and create a single optimized NESCAC volleyball conference schedule that meets schedule requirements, minimizes the total distance traveled by groups, and evenly distributes the number of home matches among teams in a group.

As discussed previously, we eventually run a parameter survey to determine the best schedule parameter settings. For now, we set an arbitrary setting (a `clustering_id` and a `home_game_constraint_combo`) and create an optimized schedule.

## Step 3 : Survey Parameters

### Survey parameters

To find out which schedule parameter settings result in the best optimized outcome, let’s run a parameter survey and build a parameter survey DataFrame. We need to iterate over the clusterings and the possible home game constraint combinations (used in the integer programming constraint to ensure home games are evenly distributed between teams).

Here's an overview of the process:

- For each clustering of the NESCAC schools
    - For each home game constraint combination
        - Create a specified number of optimized schedules
        - For schedules that are optimal with an acceptable home game range (the difference between the maximum number of home games per team and the minimum number of home games per team)
            - Analyze the schedules and average the results
    - Select the best (based on the schedule score) specified number of average results to add to the parameter survey DataFrame for this clustering

## Step 3 : Create Optimized Schedules

We are now ready to iterate over each of the schedule parameter settings from the parameter survey DataFrame to create a specified number of optimized schedules for each setting. We store the schedule information and analysis data in an optimized schedule analysis DataFrame. 

Note that we intentionally run the parameter survey and schedule creation in two separate steps. The parameter survey iterates over a large number of possible parameter setting combinations. We create only a small sample of optimized schedules for each setting to minimize computation time. In this schedule creation step, however, we run through only a fraction of the parameter setting combinations (the settings approved during the parameter survey) and accordingly can afford to create additional schedules for each setting.

Here's an overview of the process:

- For each parameter setting (clustering and home game constraint combination) from our parameter survey DataFrame
     
    - Create a specified number of optimized schedules

    - For schedules that are optimal with an acceptable home game range (the difference between the maximum number of home games per team and the minimum number of home games per team)

        - Analyze the schedules

    - Drop duplicate schedules based on the duplicate metric (default: total distance) keeping the best of the duplicates based on the schedule score

    - Select the best (based on the schedule score) specified number of schedules to add to our optimized schedule analysis DataFrame for this parameter setting

Let's create some optimized schedules.

### Consecutive Schedules

We could theoretically use the best of the single schedules year after year, but teams would play opponents at the same location each season. Each athlete would play the same opponent at the same location for all four years of her collegiate career.

To avoid location repetition, let's select groups of two and four schedules accounting for the average schedule score and match location coverage and changes.

- __Team Different Locations__

> Ideally, a pair or quad of NESCAC volleyball conference schedules would __maximize and evenly distribute team different locations__ to reduce and equitably allocate the number of matches that occur between the same teams at the same location in back-to-back schedules.

- __Team Unique Locations__

> Ideally, a pair or quad of NESCAC volleyball conference schedules would __maximize and evenly distribute team unique locations__ to maximize and equitably allocate each team's exposure to other schools in the conference.

```python
from src.configuration import *
from src.create_schedules import *
from src.figures import *
from src.optimize_schedule import *
from src.helpers import *
from src.analyze_schedule import *


# Number of best single schedules to consider for each clustering in each team
# home game (min, max, rng) group
# Single schedules then used to form schedule quads
# Recommended = 2
NUM_SINGLE_RESULTS_CONSIDER = 2

# Number of clusterings to consider in each team home game (min, max, rng) group
# Recommended = 5000 (arbitrary large number to keep all clusterings for all home game groups)
NUM_CLUSTERINGS_CONSIDER = 5000

# Number of best optimized schedule quads to keep for each home game group
# Saved in best quad optimized schedule analysis DF
# For comparison with NESCAC schedule quads
# Recommended = 1
NUM_QUAD_SCHEDULES_KEEP = 3


print("Loading necessary information ...\n")
optimized_schedule_analysis_df = pd.read_pickle(OPTIMIZED_SCHEDULE_ANALYSIS_FILENAME)
clusterings_information_df = pd.read_pickle(CLUSTERINGS_INFORMATION_FILENAME)
coordinates_df = pd.read_csv(VENUES_COORDINATES_FILENAME, index_col=0)
points = coordinates_df[['Latitude', 'Longitude']].values
id_to_name_mapping = create_venue_id_mapping_dicts_from_file(VENUES_MAPPING_FILENAME)[1]
nescac_schedule_analysis_df = pd.read_pickle(NESCAC_SCHEDULE_ANALYSIS_FILENAME)

print("Reducing number of optimized schedules ...")
reduced_optimized_schedule_analysis_df = reduce_schedules_by_cluster(
    optimized_schedule_analysis_df=optimized_schedule_analysis_df,
    num_clusterings=NUM_CLUSTERINGS_CONSIDER,
    num_to_keep=NUM_SINGLE_RESULTS_CONSIDER)

print("... Reduced from {} to {} optimized schedules ...\n".format(len(optimized_schedule_analysis_df), len(reduced_optimized_schedule_analysis_df)))

print("Creating quad optimized schedule analysis DataFrame ...")
quad_optimized_schedule_analysis_df = create_location_analysis_df(
    schedule_analysis_df=reduced_optimized_schedule_analysis_df,
    num_schedules=QUAD,
    temporal=False,
    values_to_consider=VALUES_TO_CONSIDER_PAIR_QUAD)
quad_optimized_schedule_analysis_df = create_score_columns(
    analysis_df=quad_optimized_schedule_analysis_df, 
    score_total_cost_weight=SCHEDULE_SCORE_TOTAL_COST_WEIGHT)


print("Saving quad optimized schedule location analysis DataFrame to pickle ...")
quad_optimized_schedule_analysis_df.to_pickle(QUAD_OPTIMIZED_SCHEDULE_ANALYSIS_FILENAME)
print("... Saved as \"{}\"".format(QUAD_OPTIMIZED_SCHEDULE_ANALYSIS_FILENAME))

print("Preview:")
display(quad_optimized_schedule_analysis_df.head())
print()

print("Creating best quad optimized schedule analysis DataFrame ...\n")
best_quad_optimized_schedule_analysis_df = create_best_optimized_schedule_analysis_df(
    optimized_schedule_analysis_df=quad_optimized_schedule_analysis_df,
    num_to_keep=NUM_QUAD_SCHEDULES_KEEP)
best_quad_optimized_schedule_analysis_df = create_abbreviated_schedule_id_column(
    analysis_df=best_quad_optimized_schedule_analysis_df)

print("Saving best quad optimized schedule analysis to pickle ...")
best_quad_optimized_schedule_analysis_df.to_pickle(BEST_QUAD_OPTIMIZED_SCHEDULE_ANALYSIS_FILENAME)
print("... Saved as \"{}\"".format(BEST_QUAD_OPTIMIZED_SCHEDULE_ANALYSIS_FILENAME))

print("Preview:")
display(best_quad_optimized_schedule_analysis_df.head())
print()

print("Extracting individual schedules from best quad optimized schedules ...")
schedule_indexes = list(set(sched_idx for sched_idxs in best_quad_optimized_schedule_analysis_df['Schedule Indexes'].values for sched_idx in sched_idxs))
subset_optimized_schedule_analysis_df = optimized_schedule_analysis_df.iloc[schedule_indexes]

print("Saving optimized schedules and team schedules to CSV ...\n")
save_schedules_and_team_schedules_from_df(
    venues_mapping_file=VENUES_MAPPING_FILENAME,
    analysis_df=subset_optimized_schedule_analysis_df,
    schedule_file_path=BEST_QUAD_OPTIMIZED_SCHEDULE_PATH)

print("Creating quad clusterings information DataFrame ...\n")
# Extract clustering IDs from optimized schedules
clustering_ids = subset_optimized_schedule_analysis_df['Clustering ID'].unique()
quad_clusterings_information_df = clusterings_information_df.loc[clusterings_information_df['Clustering ID'].isin(clustering_ids)]

print("Saving quad clusterings information to pickle ...")
quad_clusterings_information_df.to_pickle(QUAD_CLUSTERINGS_INFORMATION_FILENAME)
print("... Saved as \"{}\"".format(QUAD_CLUSTERINGS_INFORMATION_FILENAME))

print("Preview:")
display(quad_clusterings_information_df.head())
print()

# Iterate over quad clusterings
for idx, clustering_information in quad_clusterings_information_df.iterrows():

    print("\nPlotting clustering %s ..." % (clustering_information['Clustering ID']))
    print("Preview:")
    plot_filename = plot_and_save_clustering_on_nescac_map(
        points=points,
        clustering_information=clustering_information,
        venues_mapping_filename=VENUES_MAPPING_FILENAME,
        plot_path=QUAD_CLUSTERINGS_PLOT_PATH)
    
    print("Saving best clustering plot to PNG ..")
    print("... Saved as \"{}\"".format(plot_filename))

print("Creating quad NESCAC schedule location analysis DataFrame ...\n")
quad_nescac_schedule_analysis_df = create_location_analysis_df(
    schedule_analysis_df=nescac_schedule_analysis_df,
    num_schedules=QUAD,
    temporal=True,
    values_to_consider=VALUES_TO_CONSIDER_PAIR_QUAD)
quad_nescac_schedule_analysis_df = create_abbreviated_schedule_id_column(
    analysis_df=quad_nescac_schedule_analysis_df)
quad_nescac_schedule_analysis_df = create_score_columns(
    analysis_df=quad_nescac_schedule_analysis_df, 
    score_total_cost_weight=SCHEDULE_SCORE_TOTAL_COST_WEIGHT)


print("Saving quad NESCAC schedule location analysis DataFrame to pickle ...")
quad_nescac_schedule_analysis_df.to_pickle(QUAD_NESCAC_SCHEDULE_ANALYSIS_FILENAME)
print("... Saved as \"{}\"".format(QUAD_NESCAC_SCHEDULE_ANALYSIS_FILENAME))

print("Preview:")
display(quad_nescac_schedule_analysis_df.head())
print()
```
