### Calculate Team Percent Change Columns

```python
analysis_df = single_schedule_comparison_df.copy()

# Extract optimized schedules with correct scores
optimized_schedule_analysis_df = analysis_df[analysis_df['Schedule ID' + ''].str.contains('optimized')]

# Extract nescac schedules with correct scores
nescac_schedule_analysis_df = analysis_df[analysis_df['Schedule ID' + ''].str.contains('nescac')]

# Prevent setting with copy warning
nescac_schedule_analysis_df = nescac_schedule_analysis_df.copy()

# Add optimized change columns
for opt_idx, opt_row in optimized_schedule_analysis_df.iterrows():

    opt_abbrev_sched_id = opt_row['Abbreviated Schedule ID']

    for nes_idx, nes_row in nescac_schedule_analysis_df.iterrows():

        nescac_schedule_analysis_df.loc[nes_idx, "{} Team Duration in Hours Average Percent Change".format(opt_abbrev_sched_id)] = calculate_column_percent_change_optimized_nescac(opt_row, nes_row, 'Team Duration in Hours Average', '')

        nescac_schedule_analysis_df.loc[nes_idx, "{} Team Overnight Average Percent Change".format(opt_abbrev_sched_id)] = calculate_column_percent_change_optimized_nescac(opt_row, nes_row, 'Team Overnight Average', '')

# Recombine into one DataFrame
analysis_df = optimized_schedule_analysis_df.append(nescac_schedule_analysis_df)

single_schedule_comparison_df = analysis_df.copy()

display(single_schedule_comparison_df[single_schedule_comparison_df.columns[-4:]])
```

```python
name_to_id_mapping = create_venue_id_mapping_dicts_from_file(VENUES_MAPPING_FILENAME)[0]
name_to_id_mapping
```

```python
venue_one_way_distance_matrix = pd.read_csv(DISTANCE_MATRIX_FILENAME, index_col=0)
venue_one_way_distance_matrix
```

```python
venue_one_way_duration_matrix_hhmm = pd.read_csv(DURATION_MATRIX_HHMM_FILENAME, index_col=0)
venue_one_way_duration_matrix_hhmm
```

```python
schedule_index = 1

_ = pd.read_csv(f"results/single/optimized_schedules/{single_schedule_comparison_df['Schedule ID'][schedule_index]}.csv", index_col=0)

_['Home Game'] = _['Team 1'] == _['Location']
_['Team 1 Name'] = _['Team 1']
_['Location Name'] = _['Location']
_['Team 1'] = _['Team 1'].map(name_to_id_mapping)
_['Location'] = _['Location'].map(name_to_id_mapping)
_['Team 1 Travel Distance (mi)'] = None
_['Team 1 Travel Distance (hhmm)'] = None

for idx, row in _.iterrows():
    
    _['Team 1 Travel Distance (mi)'][idx] = venue_one_way_distance_matrix.iloc[row['Team 1'], row['Location']]
    _['Team 1 Travel Distance (hhmm)'][idx] = venue_one_way_duration_matrix_hhmm.iloc[row['Team 1'], row['Location']]


_['Distance Overnight'] = False
_['Weekend Overnight'] = False

# Maximum number of miles to drive one-way without staying for an
# overnight between Friday and Saturday matches
# Recommended = 53
# ==> max one-way distance with duration < 1 hour
weekend_overnight_threshold = 53

# Maximum number of miles to drive one-way without staying for an
# overnight
# Recommended = 153
# ==> max one-way distance with duration < 3 hours
distance_overnight_threshold = 153

# # Temporarily add a week column to the NESCAC team schedule DataFrame
# if "Week" not in _.columns:

#     _['Date'] = pd.to_datetime(_['Date'])
#     _['Week'] = _['Date'].dt.strftime('%U')

# Group team schedule DataFrame by team 1 and week
for team_week, team_week_df in _.groupby(['Team 1', 'Week']):

    team_week_df = team_week_df.reset_index(drop=False)
    curr_team = team_week[0]

    # Iterate over team 1 matches for the week
    for idx, row in team_week_df.iterrows():

        current_match = team_week_df.loc[idx]
        _idx = current_match['index']

        # Check for a distance overnight
        if is_distance_overnight(current_match, venue_one_way_distance_matrix, distance_overnight_threshold):

            _['Distance Overnight'][_idx] = True

#             distance_overnight = True

        # Check for a weekend overnight
        if idx > 0:

            previous_match = team_week_df.loc[idx - 1]

            if is_weekend_overnight(previous_match, current_match, venue_one_way_distance_matrix, weekend_overnight_threshold):

                # Count weekend overnight
                _['Weekend Overnight'][_idx] = True

#                 team_overnight_dict[curr_team] += 1

                # Remove distance overnight because have weekend overnight ==> already staying overnight
                _['Distance Overnight'][_idx] = False

#                 distance_overnight = False

                break
```
